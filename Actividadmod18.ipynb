# Elemento 2

```python
#Cuál es la relación entre edad y overall
df_relacion=df.groupby(['Age']).size().rename('Overall').reset_index().sort_values(by="Age")
```

```python
df_relacion
```

```python
#se calcula la correlación entre las variables numéricas del dataframe
df_relacion.corr()
```

```python
fig=px.imshow(df_relacion.corr(), text_auto=True, title="Correlación entre la edad y el overall")
fig.show()
```

# Elemento 3

```python
#gráfico de barras por club que indique número de jugadores
#primero se genera un dataframe basado en el club 
df_club=df.groupby('Club').size().rename('Name').reset_index().sort_values(by="Name")
```

```python
df_club
```

```python
#cambiar el nombre de la columna
df_club1=df_club.rename(columns={'Name':'Numero de jugadores'})
```

```python
df_club1
```

```python
#grafico de barras
#número de jugadores por club
#uso de plotly.bar
fig=px.bar(df_club1, x="Club",y="Numero de jugadores", title="Jugadores por club")
fig.show()
```
